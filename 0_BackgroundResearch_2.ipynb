```python
# Defining label order for the left column
label_order_left = list(df['Type'].unique())
label_order_left = label_order_left[::-1]
```

```python
# Defining label order for the right column
label_order_right = list(df['Factor'].unique())
label_order_right = label_order_right[::-1]
```

```python
# Plotting types vs factors
sankey(left=df['Type'],leftLabels=label_order_left,leftWeight=df['Type weight']*1.655,
       right=df['Factor'],rightLabels=label_order_right,rightWeight=df['Factor weight'],
       aspect=30,colorDict=colorDict_left,fontsize=12)
plt.gcf().set_size_inches(5,15)
```
