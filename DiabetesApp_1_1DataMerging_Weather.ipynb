# Historical weather records_data preprocessing

Weather data was acquired from visulacrossing.com, containing weather records per hour from 2019 to 2021. 

```python
import pandas as pd
```

### 1. Data exploration

Temperature and humidity were chosen from all parameters for further analysis due to their ease of recording in the future. Other parameters were disregarded as they are typically dependent or a consequence of either temperature or humidity.

```python
wd = pd.read_csv('WeatherKhabarovsk20190101-20211231.csv')
wd
```

### 2. Data cleaning

A data frame was created containing three columns, which were subsequently renamed Date, Temp and Humid. The values in the Date column were converted into datetime objects.

```python
#creating a new data frame
wd = wd[['datetime','temp','humidity']]
wd.head()
```

```python
#renaming columns
wd.rename(columns={'datetime':'Date','temp':'Temp','humidity':'Humid'},inplace=True)
```

```python
#converting to datetime objects
wd['Date']= pd.to_datetime(wd['Date'])
```

```python
wd.info()
```

```python
wd = wd.reset_index(drop=True)
```

```python
wd.to_csv('WeatherData.csv',index=False)
```
