```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.sankey import Sankey
```

```python
#Colors
background = '#F0F0F0'
red = '#ED462F'
pink = '#F1BBB8'
green = '#4CB866'
blue = '#1859A4'
black = '#333333'
grey = '#C6D0D2'
white = 'white'

#Matplotlib configurations
plt.rcParams['axes.facecolor'] = background
plt.rcParams['figure.facecolor'] = background
plt.rcParams['font.family'] = 'monospace'
plt.rcParams['xtick.color'] = background
plt.rcParams['ytick.color'] = background
plt.rcParams['xtick.labelcolor'] = background
plt.rcParams['ytick.labelcolor'] = background
plt.rcParams['legend.labelcolor'] = '#8C9496'
plt.rcParams['grid.color'] = background
plt.rcParams['axes.spines.left'] = False
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
plt.rcParams['axes.spines.bottom'] = False
plt.rcParams['axes.axisbelow'] = True
```

```python
# Create figure and axes
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1, 1, 1,)
fig.suptitle('\nData analysis process prior to app development: a step-by-step guide\n',
             fontsize=14,fontweight='bold')

# Create an instance of a sankey diagram
sankey = Sankey(ax=ax,
                head_angle = 135,  
                unit = None,
                margin = 0.2,
                offset = 0.15,
                shoulder = 0.0)

# Add patches
sankey.add(
    patchlabel = '1. Merging data',
    flows = [0.25,0.25,0.25,0.25,-1],
    orientations = [-1,-1,1,1,0], 
    labels = ['Health App','Dietary App','Excel Records','Weather',None],
    pathlengths = [0.25, 0.25, 0.25, 0.25,0.25],
    fc = blue,
    ec = background,
    lw = 3,
    joinstyle = 'round'
    )

sankey.add(
    patchlabel = '2. Preprocessing',
    flows = [1,-1],
    orientations = [0,0],
    pathlengths = [0.55,0.15],
    prior = 0,
    connect=(4,0),
    fc = pink,
    ec = background,
    lw = 3,
    joinstyle = 'round'
    )

sankey.add(
    patchlabel = '3. EDA',
    flows = [1,-1],
    orientations = [0,0],
    pathlengths = [0.35,0],
    prior = 1,
    connect=(1,0),
    fc=green,
    ec = background,
    lw = 3,
    joinstyle = 'round'
    )
sankey.add(
    patchlabel = '',
    flows = [1,-0.5,-0.5],
    orientations = [0,1,-1],
    pathlengths = [0,0.25,0.25],
    labels = [None,'4. Prediction analysis','5. Similarity analysis'],
    prior = 2,
    connect = (1,0),
    fc = red,
    ec = background,
    lw = 3,
    joinstyle = 'round'
    )

# Finish the diagram
diagrams = sankey.finish()
;
```

```python
# Create figure and axes
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(1, 1, 1,)
fig.suptitle('\nStep by step workflow diagram for the Diabetes App\n',
            fontsize=14,fontweight='bold')

# Create an instance of a sankey diagram
sankey = Sankey(ax=ax,
                head_angle = 135,  
                unit = None,
                margin = 0.2,
                offset = 0.15,
                shoulder = 0.0)

# Add patches
sankey.add(
    patchlabel = '1. Accessing patient data\n from Google Drive',
    flows = [1,-0.8],
    orientations = [0,0],
    pathlengths = [0.1,0.1],
    fc = blue,
    ec = background,
    lw = 3,
    joinstyle = 'round'
    )

sankey.add(
    patchlabel = '2. Data filtering and\ntrain-test split',
    flows = [0.8,-0.6,-0.2],
    orientations = [0,0,0],
    pathlengths = [1,0.1,0.1],
    fc = pink,
    ec = background,
    lw = 3,
    joinstyle = 'round',
    prior = 0,
    connect = (1,0)
    )

sankey.add(
    patchlabel = '3. Finding similar rows\n and label prediction',
    flows = [0.6,0.2,-0.8],
    orientations = [0,0,-1],
    pathlengths = [0,0,0.2],
    fc = green,
    ec = background,
    lw = 3,
    joinstyle = 'round',
    prior = 1,
    connect = (1,0)
    )

sankey.add(
    patchlabel = '4. Plotting ',
    flows = [0.8,-0.6,-0.2],
    orientations = [0,0,-1],
    pathlengths = [0,0,0.2],
    labels = [None,'Intercative\nvisualisation',None],
    fc = red,
    ec = background,
    lw = 3,
    joinstyle = 'round',
    prior = 2,
    connect = (2,0)
    )

sankey.add(
    patchlabel = '5. Modifying test data\n   using widgets ',
    flows = [0.2,-0.2],
    orientations = [0,-1],
    pathlengths = [0.1,1],
    labels = [None,'New test data'],
    fc = grey,
    ec = background,
    lw = 3,
    joinstyle = 'round',
    prior = 3,
    connect = (2,0)
    )


# Finish the diagram
diagrams = sankey.finish()

# Adjusting patchlabels
diagrams[0].text.set_position((-1,0))
diagrams[1].text.set_position((1.25,0))
diagrams[2].text.set_position((3.22,-0.25))
diagrams[3].text.set_position((2.9,-1))
diagrams[4].text.set_position((1.75,-1.7))

# Adjusting labels
diagrams[4].texts[1].set_position((1.33,-0.5))
;
```
